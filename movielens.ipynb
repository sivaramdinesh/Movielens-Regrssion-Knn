```python
df_movies['list_genre']=df_movies.Genres.str.split('|')
List_All_Genres=[]
type(df_movies['list_genre'])
```

```python
for element in df_movies['list_genre']:#list of unique genres
    List_All_Genres.extend(element)
    
```

```python
Set_List_All_Genres=set(List_All_Genres)

```

```python
Set_List_All_Genres# unique genres calculating
```

```python
df_movies=df_movies.reindex(df_movies.columns.tolist()+list(Set_List_All_Genres),axis=1)#reindexing and creating dummies for unique genres
```

```python
df_movies.shape
```

```python
df_movies.head()
df_movies=df_movies.fillna(int(0))
```

```python
for index,value in df_movies.Genres.items():# one hot encoding of genres in df_movies
    for val in Set_List_All_Genres:
        if val in value:
            df_movies[val][index]=1
        
        
    #
        #df_movies.Animation[index]=1
    
```

```python
df_movies.head(50)# output of df_movies after one hot encoding
```
